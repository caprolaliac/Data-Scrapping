```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://anilist.co/search/anime?format=TV&sort=SCORE_DESC"
```

```python
page= requests.get(url)
page
```

```python
page.content
```

```python
soup= BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
scrape_title=soup.find_all('a', class_='title')
scrape_title
```

```python
anime = []
for title in scrape_title:
    title=title.get_text().replace('\n',"")
    title=title.strip(" ")
    #title = title[:len(title)//2]
    anime.append(title)
anime
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
```

```python
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
```

```python

service = Service(driver_path)

options = Options()

driver = webdriver.Edge(service=service, options=options)
```

```python
url = 'https://anilist.co/search/anime?format=TV&sort=SCORE_DESC'
driver.get(url)
```

```python
wait = WebDriverWait(driver, 25)
chart_icon = wait.until(EC.element_to_be_clickable((By.CLASS_NAME, 'icon-wrap.chart')))
chart_icon.click()
```

```python
last_height = driver.execute_script("return document.body.scrollHeight")
```

```python
# scrolling to bottom
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(25)  # sleep time
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
driver.quit()
```

```python
scraped_title = soup.find_all('a', class_='title')
scraped_title
```

```python
for title in scraped_title:
    title_text = title.get_text().replace('\n', "").strip()
    anime.append(title_text)
anime
```

```python
scraped_score = soup.find_all('span', class_='percentage')
scraped_score
```

```python
rating=[]
for score in scraped_score:
    score_text = score.get_text().replace('\n', "").strip()
    rating.append(score_text)
rating
```

```python
scraped_des=soup.find_all('div', class_='description')
scraped_des
```

```python
plot=[]
for text in scraped_des:
    des_text = text.get_text().replace('\n', "").strip()
    plot.append(des_text)
plot
```

```python
scraped_genres=soup.find_all('div', class_='genres')
scraped_genres
```

```python
genre=[]
for genre_div in scraped_genres:
    genres = [a.get_text().strip() for a in genre_div.find_all('a', class_='genre')]
    genre.append(genres)
genre
```

------------

```python
for i, genres in enumerate(genre):
    print(f"{anime[i]}, genres: {', '.join(genres)}")
```

```python
min_length = min(len(anime), len(rating), len(plot), len(genre))

anime = anime[:min_length]
rating = rating[:min_length]
plot = plot[:min_length]
genre = genre[:min_length]
```

```python
df = pd.DataFrame({
    'Title': anime,
    'Score': rating,
    'Genres': genre,
    'Plot': plot
})
```

```python
df.to_csv('top_anime.csv', index=False)
```
